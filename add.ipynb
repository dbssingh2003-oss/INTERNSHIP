```python
# Matrix 1
A = [
    [1, 2, 3],
    [4, 5, 6]
]

# Matrix 2
B = [
    [7, 8, 9],
    [10, 11, 12]
]

# Result matrix
result = []

# Matrix addition
for i in range(len(A)):
    row = []
    for j in range(len(A[0])):
        row.append(A[i][j] + B[i][j])
    result.append(row)

# Print the result
print("Matrix Addition Result:")
for row in result:
    print(row)
```

```python
# Matrix 1
A = [
    [7, 8, 9],
    [10, 11, 12]
]

# Matrix 2
B = [
    [1, 2, 3],
    [4, 5, 6]
]

# Result matrix
result = []

# Matrix subtraction
for i in range(len(A)):
    row = []
    for j in range(len(A[0])):
        row.append(A[i][j] - B[i][j])
    result.append(row)

# Print the result
print("Matrix Subtraction Result:")
for row in result:
    print(row)
```

```python
# Matrix A (2 x 3)
A = [
    [1, 2, 3],
    [4, 5, 6]
]

# Matrix B (3 x 2)
B = [
    [7, 8],
    [9, 10],
    [11, 12]
]

# Create result matrix (2 x 2) filled with zeros
result = [[0 for j in range(len(B[0]))] for i in range(len(A))]

# Matrix multiplication
for i in range(len(A)):          # Rows of A
    for j in range(len(B[0])):   # Columns of B
        for k in range(len(B)):  # Columns of A / Rows of B
            result[i][j] += A[i][k] * B[k][j]

# Print result
print("Matrix Multiplication Result:")
for row in result:
    print(row)
```
